Importing the libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler
```

```python
#load the dataset
Admission_dataset = pd.read_csv('/content/Admission_Predict.csv')
Admission_dataset.head()
```

```python
x = Admission_dataset.iloc[:,-1].values #independent variable
y = Admission_dataset.iloc[:,1].values #dependent variable
```

```python
#features 
x = Admission_dataset.drop(["Chance of Admit ","Serial No."], axis=1)
y = Admission_dataset["Chance of Admit "]
```

```python
print(x)
```

```python
scaler = MinMaxScaler()
x = scaler.fit_transform(Admission_dataset)
```

```python
#Splitting the dataset
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Model training

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

model evaluation

```python
model.score(x_test, y_test)
```
